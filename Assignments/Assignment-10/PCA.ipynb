### Task 2: Dimensionality Reduction with PCA

```python
df1 = df.iloc[:, 1:]
df1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
df1 = sc.fit_transform(df1)
```

```python
df1
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=13)
pca_values = pca.fit_transform(df1)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var, decimals=4)*100)
var1
```

```python
plt.plot(var1, color='red', marker='o', linestyle='--')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
```

As depicted in the provided figure, six distinct components are necessary to attain a cumulative variance of 90%.

```python
pca_features = pca_values[:, :6]
```

```python
pca_df = pd.DataFrame(pca_features, columns=[f"PC{i+1}" for i in range(6)])
```

```python
pca_df
```

```python
pca_df['Type'] = df['Type']
pca_df
```

### Task 3: Clustering with Original Data

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
inertia = []
silhouette_scores = []
for k in range(2, 10):
    clf = KMeans(n_clusters=k, random_state=42)
    clf.fit(df1)
    inertia.append(clf.inertia_)
    silhouette_scores.append(silhouette_score(df1, clf.labels_))

plt.figure(figsize=(10, 5))
plt.plot(range(2,10), inertia, marker="o", linestyle="-", color="b", label="Inertia")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(2, 10), silhouette_scores, marker="o", linestyle="-", color="g", label="Silhouette Score")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Different k values")
plt.legend()
plt.show()
```

The Elbow Method suggests that the optimal number of clusters is 3, where the inertia curve starts flattening. The Silhouette Score also supports k=3 as a good choice for clustering.

```python
clf = KMeans(n_clusters=3, random_state=42)
clf.fit(df1)
```

```python
y_kmeans = clf.fit_predict(df1)
```

```python
df_kmeans = df.copy()
```

```python
df_kmeans.drop(columns=['Type'], inplace=True)
```

```python
df_kmeans['Clust'] = pd.Series(y_kmeans)
```

```python
df_kmeans.head()
```

```python
df_kmeans.plot(x='Alcohol', y='Dilution', c=df_kmeans['Clust'], kind='scatter', cmap=plt.cm.coolwarm)
plt.title('Clusters Using KMeans')
plt.show()
```

### Task 4: Clustering with PCA Data

```python
pca_df
```

```python
inertia_pca = []
silhouette_scores_pca = []
for k in range(2, 10):
    clf = KMeans(n_clusters=k, random_state=42)
    clf.fit(pca_df.iloc[:, :-1])
    inertia_pca.append(clf.inertia_)
    silhouette_scores_pca.append(silhouette_score(pca_df.iloc[:, :-1], clf.labels_))

plt.figure(figsize=(10, 5))
plt.plot(range(2,10), inertia_pca, marker="o", linestyle="-", color="b", label="Inertia")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(2, 10), silhouette_scores_pca, marker="o", linestyle="-", color="g", label="Silhouette Score")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Different k values")
plt.legend()
plt.show()
```

```python
# Using k = 3 for PCA
clf = KMeans(n_clusters=3, random_state=42)
clf.fit(pca_df.iloc[:, :-1])
```

```python
y_pca = clf.fit_predict(pca_df.iloc[:,:-1])
```

```python
pca_df['Clust'] = pd.Series(y_pca)
pca_df.head()
```

```python
pca_df.plot(x='PC1', y='PC2', c=clf.labels_, kind='scatter', cmap=plt.cm.coolwarm)
plt.title('Clusters Using KMeans')
plt.show()
```

### The silhouette score for the original data is 0.285, while the silhouette score for the PCA-transformed data is 0.345. Consequently, the higher silhouette score indicates a superior clustering performance. The PCA-transformed data marginally enhances the clustering quality.
