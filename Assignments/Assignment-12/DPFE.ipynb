- StandardScalar transforms the data to have mean = 0 and standard deciation = 1. It is used when the data is assumed to be normally distributed. It is good for Linear Regression, Logistic Regression, Support Vector Machines, Principal Component Analysis, K-Means Clustering.

##### In the above dataset, MinMaxScalar is preferred as the data is not normally distributed.

#### Encoding

- Variables with less than 5 categories is "sex"
- Variables with more than 5 categories are "'workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'native_country'"

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_cols = df_mn.select_dtypes(include='object').columns.tolist()
categorical_cols.remove('income')

# Counting unique variables in each categorical column
unique_counts = df_mn[categorical_cols].nunique()

# Applying OneHotEncoding on columns where there are less than 5 categories.
one_hot_cols = unique_counts[unique_counts < 5].index.tolist()
df_one_hot_encoded = pd.get_dummies(df_mn[one_hot_cols], prefix=one_hot_cols, dtype=int, drop_first=True)

# Applying LabelEncoding on columns where there are more than 5 categories.
label_encode_cols = unique_counts[unique_counts >= 5].index.tolist()
df_label_encoded = pd.DataFrame()

for col in label_encode_cols:
    le = LabelEncoder()
    df_label_encoded[col + '_encoded'] = le.fit_transform(df_mn[col].astype(str))

df_encoded = pd.concat([df_mn[numerical_cols], df_one_hot_encoded, df_label_encoded, df['income']], axis=1)
```

```python
pd.set_option('display.max_columns', None)
df_encoded.head()
```

```python
df_encoded.head()
```

#### One-Hot Encoding

- Pros: Doesn't assume order, ideal for non-ordinal categories.
- Cons: Increases dimensionality, sparse matrix

#### Label Encoding

- Pros: Preserves memory, simple to apply.
- Cons: Implies ordinal relationship (which may not exist), bias risk.

```python
df_encoded['capital_net_gain'] = df_encoded['capital_gain'] - df_encoded['capital_loss']
```

```python
bins = [0, 30, 50, float('inf')]
labels = ['Part-time', 'Full-time', 'Overtime']
df_encoded['hours_category'] = pd.cut(df_encoded['hours_per_week'], bins=bins, labels=labels, right=False)
df_encoded = pd.get_dummies(df_encoded, columns=['hours_category'], drop_first=True, dtype=int)
```

```python
df_encoded.head()
```

```python
df_encoded['log_capital_gain'] = np.log1p(df_encoded['capital_gain'])
```

```python
# Isolation Forest for Outlier Detection
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.01, random_state=42)
outliers = iso_forest.fit_predict(df_encoded[numerical_cols])
df_cleaned = df_encoded[outliers == 1]
```
