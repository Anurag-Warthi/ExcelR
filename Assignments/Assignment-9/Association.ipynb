```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('Online_retail.xlsx')
df
```

```python
df.columns = ['Items']
```

```python
df
```

```python
df.isna().sum()
```

```python
X = df['Items'].str.get_dummies(sep=',')
```

```python
X
```

```python
X = df['Items'].str.replace(' ', '').str.get_dummies(sep=',')
```

```python
X
```

```python
frequent_itemsets = apriori(X, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.2)
rules.sort_values('lift', ascending=False).head(10)
```

## Interview Questions

##### 1. What is lift and why is it important in Association Rules?
Lift measures how much more likely two items are to be bought together compared to if they were bought independently. It helps determine whether the association is meaningful or just due to chance. Lift helps in filtering out trivial or misleading rules by showing real dependency.

##### 2. What are Support and Confidence? How do you calculate them?

Support is how often an item occurs in the dataset. 

Support(A) = (Number of transactions containing A) / (total transactions)

    - Higher support means the item is frequently brought.
    - Low support means the item is rare.

Confidence is how often B appears when A is brought. 

Confidence (A -> B): Support( A U B) / Support(A)

    - High confidence means B is frequently bought when A is bought.

##### 3. What are some limitations or challenges of Association rules mining?
1. Too Many Rules
    - Large datasets generate thousands of rules, many of which are irrelevant.
2. Computational Complexity
    - Apriori needs to scan the dataset multiple times, making it slow for large datasets.
3. Low Support Items are Ignored
    - Rare but highly valuable associations might be ignored due to low support.
4. Misleading Confidence
    - High confidence does not always mean a strong relationship.
5. Handling Categorical & Continuous Data
    - Works best for categorical data
